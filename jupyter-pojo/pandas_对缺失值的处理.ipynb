```python
import pandas as pd
```

```python
df = pd.read_excel('./study-pandas/file/datas/student.xlsx',skiprows=2)
```

```python
df
```

### 检测空值

```python
df['分数'].isnull()
```

```python
# 判断单列不为空
df.loc[df['分数'].notnull(),:]
```

```python
# 删除空列
df.dropna(axis='columns',how='all',inplace=True)
```

```python
df
```

```python
# 删除空行
df.dropna(axis='index',how='all',inplace=True)
```

```python
df
```

```python
# 填充
df.fillna({'分数':0},inplace=True)
```

```python
df.head()
```

```python
df.loc[:,'姓名'] = df['姓名'].fillna(method='ffill')
```

```python
df
```

### 将结果集写入文件

```python
df.to_excel('./student_clean.xlsx',index=False)
```
