```python
import pandas as pd
```

```python
file = "./ant-learn-python/5分钟学会Python爬取天气数据/beijing_tianqi_2018.csv"
```

```python
df1 = pd.read_csv(file)
```

```python
df1['yWendu']=df1['yWendu'].astype(str).str.replace("℃","").astype('int32')
df1['bWendu']=df1['bWendu'].astype(str).str.replace("℃","").astype('int32')
```

```python
condition = df1['ymd'].str.startswith('2018-03')
```

```python
# 会出现 SettingWithCopyWarning错误
df1[condition]['wen_cha'] = df1['bWendu'] - df1['yWendu']
```

```python
# 查看是否成功
df1[condition].head()
```

### 1. 原因
    df1[condition]['wen_cha']是一个链式操作，先get再set，get得到的df可能是view也可能是copy。pandas发出警告
    相当于df1.get(condition).set('wen_cha')
    核心要诀：pandas的df的修改写操作，只允许在源df上进行，且一步到位

### 2. 解决方法 1
    将get + set两步操作，转为一个set操作

```python
df1.loc[condition,'wen_cha1'] = df1['bWendu'] - df1['yWendu']
```

```python
df1[condition].head()
```

### 3.解决方法2
    copy

```python
df3 = df1[condition].copy()
```

```python
df3.head()
```

```python
df3['wen_cha3'] = df1['bWendu'] - df1['yWendu']
```
