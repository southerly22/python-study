```python
import pandas as pd
df = pd.read_csv("./ant-learn-python/5分钟学会Python爬取天气数据/beijing_tianqi_2018.csv")
```

```python
df.set_index('ymd',inplace=True)
print(df)
```

```python
df.head()
```

```python
df.loc[:,"bWendu"]= df["bWendu"].str.replace("℃","").astype('int32')
df.loc[:,"yWendu"]= df["yWendu"].str.replace("℃","").astype('int32')
```

```python
df.loc[:,"wencha"] = (df["bWendu"] - df["yWendu"])
```

```python
df.head()
```

```python
def get_wd_type(df):
    if df['bWendu'] > 33:
        return '高温'
    if df['yWendu'] < -3:
        return '低温'
    return '正常'
    
# axis=0 表示index axis=1 表示列名
df.loc[:,"wd_type"] = df.apply(get_wd_type,axis=1)
```

```python
df['wd_type'].value_counts()
```

```python
# 按条件选择分组分别赋值

# 1. 先创建空列
df['wencha_type']=''

df.loc[df['bWendu']-df['yWendu']>10,'wencha_type']='温差大'
df.loc[df['bWendu']-df['yWendu']<=10,'wencha_type']='温差小'
```
